```python
from qiskit import QuantumCircuit
import numpy as np
```

```python
qc = QuantumCircuit(3)
```

```python
def phase_oracle(qc):
    qc.x(1)
    qc.ccz(0,2,1)
    qc.x(1)
    return qc.draw('mpl')
```

```python
phase_oracle(qc)
```
